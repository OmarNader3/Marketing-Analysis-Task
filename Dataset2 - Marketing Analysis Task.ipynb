```python
import pandas as pd

xls = pd. ExcelFile(r'dataset2.xlsx')
df1 = pd. read_excel(xls, 'Sales')
df2 = pd. read_excel(xls, 'Stocks')

df1
```

```python
df3 = df1.groupby(by=['sku'])['sku','Sales_Count'].sum()
df3.sort_values(by=['Sales_Count'], ascending=False)
```

```python
df4 = df2.groupby(by=['sku'])['sku','Stock'].sum()
df4.sort_values(by=['Stock'], ascending=False)
```

```python
averagepersku = df4.assign(AverageInventory = (df4['Stock'] + df3['Sales_Count']) / 2)
averagepersku.sort_values(by=['AverageInventory'], ascending=False)
```

```python
DOH = averagepersku.assign(DaysOnHand = (averagepersku['AverageInventory']/(df4['Stock']/15)))
```

```python
DOH.sort_values(by=['DaysOnHand'], ascending=False)
```
