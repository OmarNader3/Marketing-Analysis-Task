```python
import pandas as pd
df = pd.read_excel (r'dataset1.xlsx')
df.head()
```

# Number 1

```python
df2 = df.groupby(by='month').agg('count')
df2
```

```python
april = df2['week'].values[0]
may = df2['week'].values[1]
```

```python
CRR = (may/april) * 100
print("Customer Retention Rate for May is = ", CRR)
```

```python
df3 = df.groupby(by=['month', 'channel']).agg('count')
df3
```

```python
aprilapp = df3['week'].values[0]
aprilsales= df3['week'].values[1]
apriltele= df3['week'].values[2]

mayapp = df3['week'].values[3]
maysales= df3['week'].values[4]
maytele= df3['week'].values[5]
```

```python
CRR_APP = (mayapp/aprilapp) * 100
print("Customer Retention Rate for Application Channel in May is = ", CRR_APP)

CRR_AGENT = (maysales/aprilsales) * 100
print("Customer Retention Rate for Sales Agent Channel in May is = ", CRR_AGENT)

CRR_TELE = (maytele/apriltele) * 100
print("Customer Retention Rate for TeleSales Channel in May is = ", CRR_TELE)
```

```python
#Retention Rate can be enhanced by checking the channel with low retention rate and identify the problem that's causing such a low rate and come up woth solutions
#like the Tele Sales low rate, the low rate may be caused by inexperienced sales agents.
```

```python
appusersmay = df[(df.month == 5) & (df.sales_order_status == "Delivered")]
appusersmay
```

```python
appusersmay2 = appusersmay[appusersmay.groupby('retailer_id').retailer_id.transform('count') >= 2]
appusersmay2
```

```python
telemay = df[(df.month == 5) & (df.sales_order_status == "Delivered") & (df.channel == "telesales")]
telemay
```

```python
telemay2 = telemay[telemay.groupby('retailer_id').retailer_id.transform('count') >= 2]
telemay2
```

```python
telemay3 = telemay2.groupby(['retailer_id', 'sales_order_status'], as_index=False).last()
telemay3
```

```python
appmaycount = appusersmay2['retailer_id'].count()
print(appmaycount)
```

```python
telemaycount = telemay3['retailer_id'].count()
print(telemaycount)
```

```python
May = len(df[df['month'] == 5])
print(May)
```

```python
MonthlyPercentageOfAppUsersOfMay = appmaycount/May * 100
print('Percentage of App users in May = ',MonthlyPercentageOfAppUsersOfMay)
```

```python
MonthlyPercentageOfTelesalesUsersOfMay = telemaycount/May * 100 
print('Percentage of TeleSales users in May = ', MonthlyPercentageOfTelesalesUsersOfMay)
```

```python
june = df[(df.month == 6)]
june
```

```python
june15orders = june['sales_order_id'].count() 
print(june15orders)
#Since the last creation date is june 15th we'll assume that we'll start projecting from june 16th
```

```python
#Assuming that the remaining 15 days will achieve such sales on average

totaljuneorders = june15orders * 2
print('Projection of total orders in June = ',totaljuneorders)
```

```python
june = june.assign(order_price_after_discount = df['order_price_before_discount'] - df['discount'])
june
```

```python
june15sales = june['order_price_after_discount'].sum()
print(june15sales)
```

```python
totaljunesales = june15sales * 2
print('Projection of total sales in June = ',totaljunesales)
```
