```python
orders = df.groupby(['district_ar']).month.value_counts().unstack().sort_values(by=['district_ar']).fillna(0)
orders.drop(6, inplace=True, axis=1)
orders
```

```python
orders.loc[orders[4] == 0, 'percentage_of_growth_per_number_of_orders'] = (orders[5]) 
orders.loc[orders[4] > 0, 'percentage_of_growth_per_number_of_orders'] = ((orders[5] - orders[4])/orders[4])*100
orders.fillna(0,inplace=True)
orders.sort_values(by=['percentage_of_growth_per_number_of_orders'],ascending=False)
```

```python
retailers = df[df.month != 6]
retailers
```

```python
retailers.drop_duplicates(subset ="retailer_id",keep = 'first' , inplace = True)
retailers
```

```python
distinctretailers = retailers.groupby(['district_ar']).month.value_counts().unstack().sort_values(by=['district_ar']).fillna(0)
distinctretailers
```

```python
distinctretailers.loc[distinctretailers[4] == 0, 'percentage_of_growth_per_distinct_retailers'] = (distinctretailers[5]) 
distinctretailers.loc[distinctretailers[4] > 0, 'percentage_of_growth_per_distinct_retailers'] = ((distinctretailers[5] - distinctretailers[4])/distinctretailers[4])*100
distinctretailers.fillna(0,inplace=True)
distinctretailers.sort_values(by=['percentage_of_growth_per_distinct_retailers'],ascending=False)
```

```python
ddf = df.assign(order_price= df['order_price_before_discount'] - df['discount'])
ddf
```

```python
sales = ddf[(ddf.month != 6)]
sales
```

```python
sales2 = sales.groupby(['district_ar','month']).order_price.mean().reset_index()
sales2
```

```python
averagesales = sales2.groupby(['district_ar','month']).order_price.mean().unstack().reset_index().fillna(0)
averagesales.sort_values(by='district_ar',ascending=True)
```

```python
averagesales.loc[averagesales[4] == 0, 'percentage_of_growth_per_average_sales'] = (averagesales[5]) 
averagesales.loc[averagesales[4] > 0, 'percentage_of_growth_per_average_sales'] = ((averagesales[5] - averagesales[4])/averagesales[4])*100
averagesales.fillna(0,inplace=True)
averagesales.sort_values(by=['percentage_of_growth_per_average_sales'],ascending=False)
```

```python
sumofsales = sales2.groupby(['district_ar','month']).order_price.sum().unstack().reset_index().fillna(0)
sumofsales.sort_values(by='district_ar',ascending=True)
```

```python
sumofsales.loc[sumofsales[4] == 0, 'percentage_of_growth_for_revenue'] = (sumofsales[5]*100) 
sumofsales.loc[sumofsales[4] > 0, 'percentage_of_growth_for_revenue'] = ((sumofsales[5] - sumofsales[4])/sumofsales[4])*100
sumofsales.fillna(0,inplace=True)
sumofsales.sort_values(by=['percentage_of_growth_for_revenue'],ascending=False,inplace=True)
```
